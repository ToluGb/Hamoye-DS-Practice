```python
# prices
fig, ax = plt.subplots(figsize=(15,3))
ax.set_title('Box plot of the prices')
sns.boxplot(x='price', data = data)
```

Too many outliers prevent us from seeing the boxes. Hence, we have to remove the outliers using the **Interquartile range**.

```python
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['price'] >= Q1 - 1.5 * IQR) & (data['price'] <= Q3 + 1.5 * IQR)
```

```python
init_size = data.count()['id']
data = data.loc[filter]
filtered_size = data.count()['id']
print(init_size-filtered_size,'(', '{:.2f}'.format(100*(init_size-filtered_size)/init_size), '%',')', 'outliers removed from dataset')
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Distribution of the prices')
sns.distplot(data['price'], bins=30, kde=False)
```

From the plot we can see that there is a large amount of 'free' cars, that is, cars that don't cost anything. We have to remove them.

We set a threshold of $500

```python
data = data[data['price'] >= 500]
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Distribution of the prices')
sns.distplot(data['price'], bins=30, kde=False)
```

```python
# odometer
fig, ax = plt.subplots(figsize=(15,3))
sns.distplot(data['odometer'])
ax.set_title('Boxplot of the odometer')
sns.boxplot(x='odometer', data = data)
```

From the boxplot, we can see that there is a large proportion of outliers. We have to remove them by using **interquartile range** too

1. Extremly small odometer should be, as they tend to be new cars.
2. The upper threshold is set to be higher, accepting more samples.

```python
Q1 = data['odometer'].quantile(0.25)
Q3 = data['odometer'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['odometer'] <= Q3 + 3 * IQR)
init_size = data.count()['id']
data = data.loc[filter]  
filtered_size = data.count()['id']
print(init_size-filtered_size,'(', '{:.2f}'.format(100*(init_size-filtered_size)/init_size), '%',')', 'outliers removed from dataset')
```

We plot the geographical distribution of cars using the latitude and longitude coordinates

```python
fig, ax = plt.subplots(figsize=(20,15))
ax.set_title('Geographical distribution of the cars based on price')
sns.scatterplot(x= 'long', y='lat', data = data, hue = 'price', ax=ax )
```

From the plot, we can see that most of the cars are being sold in the USA. The prices don't seem to vary much geographically, which indicates that the coordinates don't have significant correlations with the prices.

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.set_title('Mean car price of each state')
data.groupby(['state']).mean()['price'].plot.bar(ax=ax)
```

We can see that the prices don't vary much in each state.

Hence, we can drop the lat, long and state columns with other irrelevant columns.
